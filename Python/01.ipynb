<b> AXES: </b>

It is the region of the image with the data space. A figure can contain many axes, but a given axes object can only be in one figure. The axes contains two (or 3) axis objects, which take care of the data limits. Each axes has a title, x-label and y-label

<b> AXIS: </b>

These are the number-line-like objects. They take care of setting the graph limits and generating the ticks and ticklabels. The location of the ticks is determined by a locator object and the ticklabel strings are formatted by a formatter. The combination of the correct locator and formatter gives very fine control over the tick locations and labels

With an arbitrary sequence of numbers, ```plt.plot()``` plots a line plot starting from zero.

So that, feeding ```plt.plot``` with a ```list``` containing the values 1, 2, 3, 4, 10: produces the following points:

(0, 1); (1, 2); (2, 3); (3, 4); (4, 10)

````python
plt.plot([1, 2, 3, 4, 10])
plt.show()
````

```plt.plot()``` expects 3 arguments in the order:

(x, y, format)

The <b> format </b> is a short hand combination of ```{color}{marker}{line}```.

As an example:

```'go-''``` produces green colored dots with solid lines

```'go'``` produces green colored dots (without any line)

Other examples are:

```'r--'*``` produces red stars with dashed lines

```'ks.'``` produces black squares with dotted line

```'bD-.'``` produces blue diamonds with dash-dot line

For a complete list ```help(plt.plot)```

<h4> Draw 2 scatterplots in different panels </h4>

we can do it by creating two separate subplots using ```plt.subplots(1, 2)```, which returns:

- the figure
- the axes (subplots) inside the figure

So, if we want to create two subplots (axes), we have to call the plot function in the respective axes, instead of ```plt```

````python
#Create figure and subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (10, 4), sharey = True, dpi = 100)

#plot
ax1.plot([1, 2, 3, 4, 5], [1, 2, 3, 4, 10], 'go')
ax2.plot([1, 2, 3, 4, 5], [2, 3, 4, 5, 11], 'b*')

#title, X and Y labels, X and Y lim
ax1.set_title('Scatterplot Greendots')
ax2.set_title('Scatterplot Greendots')
ax1.set_xlabel('X'), ax2.set_xlabel('X')
ax1.set_ylabel('Y'), ax2.set_ylabel('Y')
ax1.set_xlim(0, 6), ax2.set_xlim(0, 6)
ax1.set_ylim(0, 12), ax2.set_ylim(0, 12)

#ax2.yaxis.set_ticks_position('none')
plt.tight_layout()
plt.show()
````

<b> to avoid duplicating y ticks: </b>

```plt.tight_layout()```

<b> to share Y axis between 2 subplots: </b>

```sharey = True```

```dpi = 100``` is used to increase the number of dots per inch and make the plot look more sharp and clear

ax1 and ax2 objects, like ```plt``` has equivalent ```set_title```, ```set_xlabel``` and ```set_ylabel``` functions:

- ```plt.xlabel()``` -> ```ax.set_xlabel()```
- ```plt.ylabel()``` -> ```ax.set_ylabel()```
- ```plt.xlim()``` -> ```ax.set_xlim()```
- ```plt.ylim()``` -> ```ax.set_ylim()```
- ```plt.title()``` -> ```ax.set_title()```

As an alternative, we can set multiple things in one, as follows:

````python
ax1.set(title = 'Scatterplot Greendots',
       xlabel = 'X', ylabel = 'Y',
        xlim = (0, 6), ylim = (0, 12))
ax2.set(title = 'Scatterplot Bluestars',
       xlabel = 'X', ylabel = 'Y',
       xlim = (0, 6), ylim = (0, 12))
````

<h4> Object Oriented Syntax vs. Matlab like Syntax </h4>

A known problem with learning matplotlib is it has two coding interfaces:

- matlab like syntax
- object oriented syntax

```plt``` keeps track of what the current axes is. So, whatever you draw with ```plt.{anything}``` will reflect only on the current subplot.

Practically, the main difference between the 2 syntaxes is that:
- in matlab-like syntax all plotting is done using ```plt``` methods instead of the respective ```axes``` method as in the object oriented syntax

How to recreate the above multi-subplots figure using matlab-like syntax?

The general procedure is to create a subplot at a time using ```plt.subplot()``` or ```plt.add_subplot()```

Then, call ```plt.plot()``` or ```plt.{anything}``` to modify that specific subplot

Whatever method you call using ```plt``` will be drawn in the current axes

````python
#open figure device
plt.figure(figsize = (10, 4), dpi = 100)

#left hand side plot
plt.subplot(1, 2, 1) #(nRows, nColumns, axes number to plot)
plt.plot([1, 2, 3, 4, 5], [1, 2, 3, 4, 10], 'go')
plt.title('Scatterplot Greendots')
plt.xlabel('X'), plt.ylabel('Y')
plt.xlim(0, 6), plt.ylim(0, 12)

#right hand side plot
plt.subplot(1, 2, 2)
plt.plot([1, 2, 3, 4, 5], [2, 3, 4, 5, 11], 'b*')
plt.xlabel('X'), plt.ylabel('Y')
plt.xlim(0, 6), plt.ylim(0, 12)

plt.show()
````

In ```plt.subplot(1, 2, 1)``` the first two values specifies the number of <b>rows</b> and <b>columns</b>, the third parameter specifies the <b>position</b> of the current subplot.

The subsequent ```plt``` functions, will always draw on this current subplot.

You can get a reference to the current subplot axes with ```plt.gca()``` and the current figure with ```plt.gcf()```.

Likewise ```plt.cla()``` and ```plt.clf()``` will clear the current axes and figure respectively.

Since ```plt.subplots``` returns all the axes as separate objects, you can avoid writing repetitive code by looping through the axes.

Remember:

```plt.plot()``` or ```plt.{anything}``` will always act on the plot in the current axes, whereas ```ax.{anything}``` will modify the plot inside that specific ax

````python
#Draw multiple plots using for-loops using object oriented syntax

import numpy as np
from numpy.random import seed, randint
seed(100)

#Create Figure and Subplots

fig, axes = plt.subplots(2, 2, figsize = (10, 8),
                        sharex = True, sharey = True, dpi = 80)

#Define the colors and markers to use

colors = {0 : 'g', 1 : 'b', 2 : 'r', 3 : 'y'}
markers = {0 : 'o', 1 : 'x', 2 : '*', 3 : 'p'}

#Plot each axes

for i, ax in enumerate(axes.ravel()):
    ax.plot(sorted(randint(0, 10, 10)),
           sorted(randint(0, 10, 10)),
           marker = markers[i],
           color = colors[i])
    ax.set_title(f'Ax: {str(i)}')
    ax.yaxis.set_ticks_position('none')
ax.set_ylim(0, 10), ax.set_xlim(0, 10)
    
plt.suptitle('Four Subplots in One Figure',
            verticalalignment = 'bottom', fontsize = 16)

plt.tight_layout()
plt.show()
````

````python
for i, j in enumerate(axes.ravel()):
    print(i, j)
````

```ax.yaxis.set_ticks_position('none')``` turn off the Y-axis ticks

```plt.suptitle()``` add a main title at figure level

```title.plt.title()``` would have done the same for the current subplot (axes)

```verticalalignment = 'bottom'``` denotes the hingepoint should be at the bottom of the title text, so that the main title is pushed slightly upwards.

But it is equal to ```va = 'bottom'```

<h4> Modify the Axis Ticks Positions and Labels </h4>

There are 3 basic things you will ever need in matplotlib when it comes to manipulating axis ticks:

- How to control the position and ticks labels
- How to control which axis's ticks (top/bottom/left/right) should be displayed using ```plt.ticks_params()```
- Functional formatting of tick labels

If you are using ax syntax, you can use ```ax.set_xticks()``` and ```ax.set_xtickslabels()``` to set the positions and label texts respectively.

If you are using the plt syntax, you can set both the positions as well as the label text in one call using the ```plt.xticks()```

```plt.xticks``` takes the ticks and labels as required parameters but you can also adjust the label's fontsize, rotation, 'ha' and 'va' of the hinge points on the labels, like done below:

````python
from matplotlib.ticker import FuncFormatter

def rad_to_degrees(x, pos):
    """
    converts radians to degrees
    """
    return round(x * 57.2985, 2)

plt.figure(figsize = (12, 7), dpi = 80)
X = np.linspace(0, 2 * np.pi, 1000) #evenly spaced numbers between 0 and 2 * np.pi, creating 1000 intervals
plt.plot(X, np.sin(X)) #vectorization, this way we avoid cycling
plt.plot(X, np.cos(X)) #vectorization

#1. Adjust x axis Ticks
plt.xticks(ticks = np.arange(0, 440/57.2985, 90/57.2985), #same as range, but return an ndarray instead of a list
          fontsize = 12, rotation = 30, ha = 'center', va = 'top') #1 radian = 57.2985 degrees
#ha = horizontal alignment
#va = vertical alignment


#2. Ticks Parameters
plt.tick_params(axis = 'both',            #ticks on both axis
                   bottom = True,
                   top = True,
                   left = True,
                   right = True,
                   direction = 'in',
                   which = 'major',
                   grid_color = 'blue')
#tick on all sides of the plot
#direction of line pointing 'in' = inside
#which element of the grid to draw: only in correspondence of major grid lines
#color of grid: blue

#3. Format tick labels to convert radians to degrees
formatter = FuncFormatter(rad_to_degrees)         #feed a format function to FuncFormatter
plt.gca().xaxis.set_major_formatter(formatter)    #call the graphical current axes and on xaxes set the major formatter

plt.grid(linestyle = '--', linewidth = 0.5, alpha = 0.15)
plt.title('Sine and Cosine Waves\n(Notice the ticks are on all 4 sides pointing inwards, radians converted to degrees in x axis)')
plt.show()
````

<h4> Legend Customization </h4>

The most common way to do it is to define the label parameter for each of the plots and finally call ```plt.legend()```

However, sometimes you might want to construct the legend on your own. In that case you need to pass the plot items you want to draw the legend for, and the legend text as parameters to ```plt.legend()``` in the following format:

```plt.legend((line1, line2, line3), ('label1', 'label2', 'label3'))```

````python
plt.figure(figsize = (10, 7), dpi = 80)
X = np.linspace(0, 2 * np.pi, 1000)

sine = plt.plot(X, np.sin(X))
cosine = plt.plot(X, np.cos(X))
sine_2 = plt.plot(X, np.sin(X + .5))
cosine_2 = plt.plot(X, np.cos(X + .5))

plt.gca().set(ylim = (-1.25, 1.5), xlim = (-.5, 7))
plt.title('Custom Legend Example', fontsize = 18)

#modify legend

plt.legend([sine[0], cosine[0], sine_2[0], cosine_2[0]], #plot items
          ['sine curve', 'cosine curve', 'sine curve 2', 'cosine curve 2'],
          frameon = True,  #legend border
          framealpha = 1,  #transparency of border
          ncol = 2,        #num columns
          shadow = True,   #shadow on
          borderpad = 1,   #thickness of border
          title = 'Sines and Cosines') #title

plt.show()
````

<h4> Add texts, arrows and annotations</h4>

```plt.text``` and ```plt.annotate``` adds the texts and annotations respectively.

If you have to plot multiple texts you need to call ```plt.text()``` as many times, typically in a for-loop.

Let's annotate the peaks and troughs adding arrowprops and a box for the text

````python
plt.figure(figsize = (14, 7), dpi = 60)
X = np.linspace(0, 8 * np.pi, 1000)
sine = plt.plot(X, np.sin(X), color = 'tab:blue')

#1. annotate with arrow props and bbox
plt.annotate('Peaks', xy = (90/57.2985, 1.0), xytext = (90/57.2985, 1.5),
            bbox = dict(boxstyle = 'square', fc = 'green', linewidth = 0.1),
            arrowprops = dict(facecolor = 'green', shrink = 0.01, width = 0.1),
            fontsize = 12, color = 'white', ha = 'center')

#2. Texts at Peaks and Troughs
for angle in [440, 810, 1170]:
    plt.text(angle/57.2985, 1.05, str(angle) + '\ndegrees', transform = plt.gca().transData,
            ha = 'center', color = 'green')

for angle in [270, 630, 990, 1350]:
    plt.text(angle/57.2985, -1.3, str(angle) + '\ndegres', transform = plt.gca().transData,
            ha = 'center', color = 'red')

plt.gca().set(ylim = (-2, 2), xlim = (-.5, 26))
plt.title('Annotations and Texts Example', fontsize = 18)
plt.show()
````

Notice that all the text plotted above was in relation to the data.

That is, the x and y position in th e ```plt.text()``` corresponds to the values along the x and y axes.

However, sometimes you might work with data of different scales on different subplots and you want to write the texts in the same position on all the subplots.

In such cases, you can specify the x and y values in relation to the axes. You can do this by setting ```transform = ax.transData```.

The lower left corner of the axes has (x, y) = (0, 0) and the top right corner will correspond to (1, 1).

The plot below shows the position of texts for the same values of (x, y) = (0.5, 0.02) with respect to ```Data(transData)```, ```Axes(transAxes)``` and ```figure(transFigure)```, respectively:

````python
#Texts, Arrows, and Annotations Example
plt.figure(figsize = (14, 7), dpi = 80)
X = np.linspace(0, 8 * np.pi, 1000)

#Text relative to DATA
plt.text(0.5, 0.02, 'Text relative to the DATA centered at : (0.5, 0.02)', transform = plt.gca().transData, fontsize = 14, ha = 'center', color = 'blue')

#Text Relative to AXES
plt.text(0.5, 0.02, 'Text relative to the AXES centered at : (0.5, 0.02)', transform = plt.gca().transAxes, fontsize = 14, ha = 'center', color = 'blue')

#Text Relative to FIGURE
plt.text(0.5, 0.02, 'Text relative to the FIGURE centered at : (0.5, 0.02)', transform = plt.gcf().transFigure, fontsize = 14, ha = 'center', color = 'blue')

plt.gca().set(ylim = (-2, 2), xlim = (0, 2))
plt.title('Placing Texts Relative to Data, Axes and Figure', fontsize = 18)
plt.show()
````
